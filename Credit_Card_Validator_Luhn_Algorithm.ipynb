```python
def validate_luhn(card_number):
    # Sanitize the input by removing spaces and dashes
    card_number = card_number.replace(" ", "").replace("-", "")

    # Ensure the input contains only numerical digits
    if not card_number.isdigit():
        return False

    total_sum = 0
    # Reverse the string to process from right to left
    reversed_card = card_number[::-1]

    for index, char in enumerate(reversed_card):
        digit = int(char)

        # Double every second digit (index 1, 3, 5, etc.)
        if index % 2 == 1:
            digit *= 2
            # If doubling results in a two-digit number, subtract 9
            if digit > 9:
                digit -= 9

        total_sum += digit

    # The card is valid if the total sum is a multiple of 10
    return total_sum % 10 == 0
```

```python
# --- Google Colab Interactive Execution ---
if __name__ == "__main__":
    print("💳 Luhn Algorithm Validator 💳")
    user_input = input("Enter a credit card number to test: ")

    if validate_luhn(user_input):
        print("\n✅ VALID: The number passes Luhn's Check.")
    else:
        print("\n❌ INVALID: The number failed the checksum or contains invalid characters.")
```
